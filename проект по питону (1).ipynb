```python
quiz = {
    "1. Какой фильм КВМ начал первую фазу под названием Сага бесконечности?":
        ["A. Человек-паук", "B. Железный человек", "C. Невероятный халк", "B"],
 
    "2. В каком из миров живет Тор?":
        ["A. Ётунхейм", "B. Мидгард", "C. Асгард", "C"],
 
    "3. Из чего сделан щит Капитана Америки?":
        ["A. титан", "B. вибраниум", "C. вольфрам", "B"],
 
    "4. Как зовут Черную Пантеру?":
        ["A. T'Чалла", "B. М'Баку", "C. Н'Джобу", "A"],
 
    "5. Кто убил родителей Тони Старка?":
        ["A. зимний солдат", "B. танос", "C. локи", "A"],
 
    "6. Какая организация сделала Баки Барнса своим оружием?":
        ["A. щит", "B. гидра", "C. зодиак", "B"],
 
    "7. Кто сыграл первого человека-паука?":
        ["A. Тоби Магуайер", "B. Эндрю Гарфилд", "C. Том Холланд", "A"],
 
    "8. Кому отдал свой щит Стив Роджерс?":
        ["A. Баки Барнс", "B. Сэм Уилсон", "C. Брюс Беннер", "D. Mark Ruffalo", "B"],
 
    "9. Где родилась Наташа Романофф??":
        ["A. Сталинград", "B. Одесса", "C. Ленинград", "D. Trey Parker", "A"],
 
    "10. Сколько всего камней бесконечности?":
        ["A. 6", "B. 7", "C. 8", "A"],
 
    "11. Как называется компьютерная система Тони Старка?":
        ["A. Джеффри", "B. Дэрил", "C. Джарвис", "C"]
}
```

```python
import time
from questions import quiz
 
quiz = quiz
score = 0
answers = []
name = input("Привет! Как вас зовут?\n")
 
 
def greetings(): 
    print(
        f"Добро пожаловать, {name}! Сейчас вам предстоит пройти квиз 'Как хорошо вы знаете киновселенную Марвел?'"
        f"Удачи!\n") # Функция принимает имя пользователя и выводит приветственное сообщение
    time.sleep(2) # Функция использует модуль time для паузы в 2 секунды перед началом квиза
    print(
        "Давайте начнём, если вам надоест и Вы захотите прекратить введите -1, но учтите, "
        "что результаты не сохранятся\n") 
    # Функция выводит предупреждение, что результаты не сохранятся, если пользователь прервет прохождение квиза
 
 
greetings()
 
 
def ask_question(question, options, correct_answer): 
    # Функция принимает вопрос, список вариантов ответов и правильный ответ на вопрос
    print(question) # Функция выводит вопрос и список вариантов ответов для пользователя
    for i in range(len(options) - 1):
        print(' '.join(options))
        answer = input("Ваш ответ: ") # Функция сохраняет ответ пользователя и проверяет его на правильность
        answers.append(answer)
        while answer not in ['-1', 'a', 'b', 'c']:
            answer = input("Ты вводишь что-то не то\n")
            # Если пользователь вводит неверный ответ, функция запрашивает ввод до тех пор, пока не будет получен верный ответ
 
        return check_answer(answer, correct_answer) #Функция возвращает True, если ответ верный, и False, если неверный
 
 
def check_answer(answer, correct_answer): # Функция принимает ответ пользователя и правильный ответ на вопрос
    if answer == '-1':
        exit(print("До свидания! Жду нашей повторной встречи!")) 
        # Если пользователь решил прервать прохождение квиза, программа завершается
    else:
        if answer.lower() == correct_answer.lower(): 
        # Функция проверяет, совпадает ли ответ пользователя с правильным ответом на вопрос, игнорируя регистр букв
            print("Правильно!\n")
            return True
        else:
            print("Неправильно :( \n")
            return False
        # Если ответ верный, функция возвращает True, в противном случае возвращает False
 
 
def run_quiz(): 
# Функция запускает квиз, перебирая все вопросы из словаря в модуле questions и вызывая функцию ask_question() для каждого вопроса
    global score
    for q, a in quiz.items():
        if ask_question(q, a[:3], a[-1]): 
        # Если ответ на вопрос верный, то результат пользователя увеличивается на 1, иначе пользователю предлагается повторить ответ
            score += 1
 
        else:
            answers.pop()
            print('Попробуй еще раз!\n')
            if ask_question('', a[:3], a[-1]):
                pass
            else:
                score -= 1
 
 
run_quiz()
print(f'Спасибо за прохождение,{name}\n!!!') 
 
 
def write_results(): # Функция открывает файл для записи результатов пользователя и правильных ответов на все вопросы из словаря
    global name
    k = -1
    global answers
    with open("quiz_results.txt", "w") as file: 
    # Функция записывает имя пользователя, все вопросы и ответы пользователя на них, правильные ответы и результат пользователя в текстовый файл
        file.write(f"{name} Результаты:\n\n")
        for question, answer in quiz.items():
            k += 1
            file.write(f"{question}\n")
            file.write(f"Ваш ответ : {answers[k].upper()} \n")
            file.write(f"Правильный ответ : {answer[-1]}\n")
        if score == len(quiz):
            file.write(f"Ваш результат: {score}/{len(quiz)}\n Это максимум! Поздравляю!")
        if 8 < score < 11:
            file.write(f"Ваш результат: {score}/{len(quiz)}\n Супер! Вы почти идеально знаете кионовселенную Марвел")
        if 5 < score < 7:
            file.write(f"Ваш результат: {score}/{len(quiz)}\n Больше половины, уже неплохо, но кажется, "
                       f"что стоит пересмотреть парочку фильмов!")
        if score < 3:
            file.write(
                f"Ваш результат: {score}/{len(quiz)}\n Кажется, что вам еще много предстоит узнать об этой замечательной"
                f" киновселенной!")
 
 
write_results() # Функция закрывает файл после записи
 
```
